```python
# Read MRM lipid database
import pandas as pd
mrm_database = pd.read_excel('Lipid_Database.xlsx')

# Standardize column names (remove spaces and rename for consistency)
mrm_database.columns = mrm_database.columns.str.replace(" ", "_")  # Replace spaces with underscores
mrm_database = mrm_database.rename(columns={
    "Parent_Ion": "Parent_Ion",
    "Product_Ion": "Product_Ion",
    "Compound_Name": "Lipid"  # Rename 'Compound Name' to 'Lipid'
})

# Print column names to verify
print("MRM Database Columns:", mrm_database.columns)
#save to xlsx
mrm_database.to_excel('mrm_database.xlsx', index=False)
mrm_database
```
